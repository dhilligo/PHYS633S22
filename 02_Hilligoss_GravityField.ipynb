We know that the mass of a spherical shell of density $\rho(r)$ will be $dM_r=4\pi ρ r^2 dr$. Therefore we can find the mass enclosed $M_r$ with this chain of equation:

> <font color='blue'>(Note that you can insert a file in the markdown as described in the reference cell about images at the top of this notebook</font>)

<img src="https://github.com/veropetit/PHYS633-S2022/blob/main/L02-GravityField/02-datafile.jpeg?raw=true" width="500"> 

We can then put equation 3 and 2 in equation 1 and isolate $M_r$, and pull out then density from the integral as it is constant with $r$, such that <font color='blue'>
> (This is what I mean by "I don't need to see the purely algebra steps")</font>:

$$ M_r = 4\pi \rho \int_0^r r^{'2} dr' $$

$$ = 4\pi \rho  \left[\frac{r^{'3}}{3}\right]_0^r = \frac{4\pi r^3 \rho}{3}$$

We can normalize by the total mass of the star, which can be thought of as $M_\star = M_r(r=R_\star) = 4\pi\rho R_\star^3/3$

Therefore:


$$\frac{M_r}{M_\star} = \left(\frac{r}{R_\star}\right)^3,$$

and

$$ \frac{g}{g_\star} = \frac{M_r}{M_\star}\left(\frac{R_\star}{r}\right)^2 $$
(with the latter valid for any density profile).

The first set of calculations shows how to calculate the scaled mass for a non-constant density. Very similar to the calculation for a star with constant density, we can use integration to solve for the scaled mass. The only difference now is that the density is not a constant (dependent on r). Therefore, we need to adjust our integration to account for the increased dependence on r.
The second calculation calculates the scaled gravity $\frac{g}{g_\star}$. To begin, we know that gravitational acceleration $g(r)=\frac{GM_r}{r^2}$. To scale this, we simply choose a location where the gravitational acceleration is known (namely, the surface of the star). So, the scaled density is $\frac{g}{g_\star}$ =$\frac{GM_r/r^2}{GM_\star/R_\star^2}$. As shown, this ultimately simplifies to $\frac{g}{g_\star}$=$(M/M_\star)(R_\star/r)^2$. In the case of a constant density, we can reduce the scaled mass, $\frac{M}{M_\star}$, to simply $(\frac{r}{R_\star})^3$. But the more general form of the scaled gravity is found by $\frac{g}{g_\star}$=$(M/M_\star)(R_\star/r)^2$

````python
fig, ax = plt.subplots(1,2, figsize=(11,6))
# The subplot routine creates a figure object (in the variable "fig"), which contains
# and an array (here 2 plots) of plotting windows called "axes" (in the variable "ax")

plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'font.family': 'Times New Roman'})
# Set the font and font size for the whole figure.

for item in ax:
    item.set_xlabel(r'$r/R_\star$') # the "r" is to say there is latex formatted strings.
    item.set_xlim(0,1)
    item.set_ylabel(r'$\rho/\rho_c, M_r/M_\star, g/g_\star$') 
    item.set_ylim(0,1.5)
# ax is an vector of plot objects.
# we use a loop to set the title and limits of all plots to the same values

ax[0].set_title('Constant density')
ax[1].set_title(r'$\rho = \rho_c(1- r/R_\star)$')

#-----------------------------------------------------
#-----------------------------------------------------

#-----------------------------------------------------
# Part to do in class
#----

##################
# constant density
##################

start=0
stop=1
num_step=100

r_rs=np.linspace(start,stop,num_step)
rho0=1.0
rho1=rho0*(1-r_rs)
M_Ms=r_rs**3

ax[0].plot(r_rs,M_Ms,label=r'$M/M_\star$')
ax[0].hlines(rho0,0,1,label=r'$\rho/\rho_c$')
####################
# Decreasing density at home
####################
M_Ms1=4*(r_rs**3)-3*(r_rs**4)

ax[1].plot(r_rs,M_Ms1,label=r'$M/M_\star$')
ax[1].plot(r_rs,rho1,label=r'$\rho/\rho_c$')

#-----------------------------------------------------
# At home, add the curves for the g(s). 
# Analytically find at which r/Rstar is g at its maximum, and add a vertical line at this position.
#----
g_gs=M_Ms*((1/r_rs)**2)
ax[0].plot(r_rs,g_gs,label=r'$g/g_\star$')

g_gs1=M_Ms1*((1/r_rs)**2)
ax[1].plot(r_rs,g_gs1,label='$g/g_\star$')
ax[1].vlines(2/3,0,1.5,label='max g')
#-----------------------------------------------------
#-----------------------------------------------------
#------------------------------------------------------
for item in ax:
    item.legend(loc=0, fontsize=15)
plt.tight_layout()
# arrange the plot nicely
````

The above results show that in the case of a constant density, the greatest scaled mass and scaled gravity lie at the surface of the star. In the case of the scaled gravity, both the $M_r$ and $r^2$ contribute to the increased scaled gravity. Since we see that the scaled gravity is, in fact, increasing as shown in the graph, we can reasonably assume that the $M_r$ term dominates.

In the case of the decreasing density profile, we now see that the scaled mass increases with radius. However, the scaled gravity reaches a maximum point at roughly $r=\frac{2}{3}R_\star$. This decreasing density profile makes more physical sense but still does not seem correct because when $r<\frac{2}{3}R_\star$, the scaled gravity is still increasing in value.
